```python
import pandas as pd

titanic_df = pd.read_csv("/kaggle/input/titanic-csv/titanic.csv")

# 1. Gives number of rows, missing value counts, and datatypes  
titanic_df.info() 
```

This dataset has 891 instances and 12 attributes  
b)  
Age, Cabin, and Embarked are all missing values. Age has only 714 out of 891, Cabin has 204 out of 891, and Embarked has 889 out of 891  
c)  
For Embarked, I would fill in the missing values with the mode since only two values are missing  
For Age, I would estimate the missing values based on class and sex  
For Cabin, I would turn it into a simple boolean value, HasCabin, since estimating so many values would introduce noise

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
mosaic(titanic_df, ['Survived', 'Sex'])
plt.show()
```

e)  

```python
# Contingency table
ct = pd.crosstab(titanic_df['Survived'], titanic_df['Sex'])

# print(ct.index)    # shows row labels (Survived values)
# print(ct.columns)  # shows column labels (Sex categories)

# P(Survived | Male)
p_survived_given_male = ct.loc[1, 'male'] / ct['male'].sum()

# # P(Survived | Female)
p_survived_given_female = ct.loc[1, 'female'] / ct['female'].sum()

print(f"p_survived_given_male: {p_survived_given_male}")
print(f"p_survived_given_female: {p_survived_given_female}")
```
